```python
import pandas as pd
import re
```

```python
df = pd.read_csv("/content/drive/MyDrive/emails.csv")
```

```python
pd.set_option('display.max_colwidth',None)
```

```python
df.head()
```

```python
# Define regular expressions for extracting information
from_pattern = re.compile(r'From: (.+)')
to_pattern = re.compile(r'To: (.+)')
content_pattern = re.compile(r'Content-Type:.*?(\n\n(.+))', re.DOTALL)

# Extract information using regular expressions
df['From'] = df['message'].apply(lambda x: from_pattern.search(x).group(1) if from_pattern.search(x) else '')
df['To'] = df['message'].apply(lambda x: to_pattern.search(x).group(1) if to_pattern.search(x) else '')
df['Content'] = df['message'].apply(lambda x: content_pattern.search(x).group(2).strip() if content_pattern.search(x) else '')
```

```python
df.head(2)
```

```python
df['Content'] = df['Content'].str.replace('\n','')
```

```python
columns_to_delete = ['file', 'message']

# Check if the columns exist in the DataFrame before attempting to delete
columns_exist = all(col in df.columns for col in columns_to_delete)

# Drop the specified columns
df.drop(columns=[col for col in columns_to_delete if col in df.columns], inplace=True)

if columns_exist:
    print(f"The columns {columns_to_delete} have been deleted.")
else:
    print(f"The columns {columns_to_delete} do not exist in the DataFrame.")
```
